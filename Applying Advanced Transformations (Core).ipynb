```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json

```

```python
df_info = pd.read_csv('superhero_info - superhero_info (1).csv')
df_powers = pd.read_csv('superhero_powers - superhero_powers.csv')
```

```python
## save the 2 new columns into the dataframe
df_info[["Hero","Publisher"]] = df_info['Hero|Publisher'].str.split('|',expand=True)
df_info.head(2)
```

```python
## drop the original column 
df_info = df_info.drop(columns=['Hero|Publisher'])
```

```python
df_info.head()
```

```python
## examining a single value from the coordinates col
coord = df_info.loc[0,"Measurements"]
print(type(coord))
coord
```

```python
coord = coord.replace("'",'"')
coord
```

```python
## now we can use json.loads
fixed_coord = json.loads(coord)
print(type(fixed_coord))
fixed_coord
```

```python
## use .str.replace to replace all single quotes
df_info['Measurements'] = df_info['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
df_info['Measurements'] = df_info['Measurements'].apply(json.loads)
df_info['Measurements'].head()
```

```python
df_info= pd.concat([df_info,df_info['Measurements'].apply(pd.Series)],axis = 1)
df_info.head()
```

```python
df_info = df_info.drop(columns= ['Measurements'])
df_info.head()
```

```python
replace_cm = ['cm']
replace_kg = ['kg']

for char in replace_cm:
    df_info['Height'] = df_info['Height'].str.replace(char,'',regex = False)
    
df_info['Height'].head()
```

```python
replace_kg = ['kg']

for char in replace_kg:
    df_info['Weight'] = df_info['Weight'].str.replace(char,'',regex = False)

df_info['Weight'].head()
```

```python
df_info['Height'] = pd.to_numeric(df_info['Height'], errors='coerce')
```

```python
df_info['Weight'] = pd.to_numeric(df_info['Weight'], errors='coerce')
```

```python
df_powers['powers_split'] = df_powers['Powers'].str.split(',')
```

```python
## exploding the column of lists
exploded = df_powers.explode('powers_split')
exploded[['hero_names','Powers','powers_split']].head(5)
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df_powers[col] = df_powers['Powers'].str.contains(col)
df_powers.head()
```

```python
# drop transactions clumns
df_powers = df_powers.drop(columns=['Powers','powers_split'])
```
